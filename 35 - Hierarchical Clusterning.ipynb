```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Iris.csv')
data.head()
```

```python
data.drop(columns=['Species','Id'],inplace=True)
```

```python
data.head()
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(data,method='single',metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering( n_clusters=2,linkage='single')
data['Predict'] = ac.fit_predict(data)
```

```python
data.head()
```

```python
sns.pairplot(data=data,hue='Predict')
plt.show()
```
